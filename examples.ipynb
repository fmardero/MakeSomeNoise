# MAKE SOME NOISE
The library collects some of the most common data augumentation tecniques.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from makesomenoise import *
```

## Original Image

```python
img = mpimg.imread('armchair.jpg')

plt.imshow(img)
plt.title('Original Image')
plt.show()
```

## Random Crop

```python
noisy = MakeSomeNoise(img)
noisy.randomCropping(starting_prop = [1/2, 2/3.])

plt.imshow(noisy.image)
plt.title('Cropped Image')
plt.show()
```

## Random Cropping & Rotation

```python
noisy = MakeSomeNoise(img)
noisy.randomRotation(max_rot = 10)
noisy.randomCropping()

plt.imshow(noisy.image)
plt.title('Cropped and Rotated Image')
plt.show()
```

## Random Brightness Change

```python
noisy = MakeSomeNoise(img)
noisy.randomBrightness()

plt.imshow(noisy.image)
plt.title('Brighter/Darker Image')
plt.show()
```

## Random Noise Summation

```python
noisy = MakeSomeNoise(img)
noisy.randomNoise()

plt.imshow(noisy.image)
plt.title('Noisy Image')
plt.show()
```

## Pad/Rescale Image

```python
noisy = MakeSomeNoise(img)
noisy.padImage((150,150))

plt.imshow(noisy.image)
plt.title('Rescaled and Padded Image')
plt.show()

######################################

noisy = MakeSomeNoise(img)
noisy.padImage((1000,1000))

plt.imshow(noisy.image)
plt.title('Padded Image')
plt.show()
```

## Grayscale Image

```python
noisy = MakeSomeNoise(img)
noisy.grayScale()

plt.imshow(noisy.image, cmap = 'gray', interpolation = 'bicubic')
plt.title('Grayscale Image')
plt.show()
```

## Random Light Source

```python
noisy = MakeSomeNoise(img)
noisy.randomLightSource()

plt.imshow(noisy.image)
plt.title('Image with a Random Light Source')
plt.show()
```
